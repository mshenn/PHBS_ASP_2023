```python
import pyfeng as pf
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Heston

strike = np.linspace(0.7, 1.3, 100) * 100
sigma, vov, mr, rho, theta, texp, texp2, spot = 0.2556 ** 2, 0.9288, 3.8, -0.7829, 0.3095 ** 2, 0.25, 0.5, 100
m = pf.HestonFft(sigma, vov=vov, mr=mr, rho=rho, theta=theta)
p = m.price(strike, spot, texp)
p2 = m.price(strike, spot, texp2)

demovol = pf.Bsm(sigma=1)   # initialize a Bsm
impvol = demovol.impvol(p, strike, 100, texp)
impvol2 = demovol.impvol(p2, strike, 100, texp2)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].plot(strike / spot, impvol)
ax[1].plot(strike / spot, impvol2)
```
